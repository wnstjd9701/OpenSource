```python
# 2번 문제
n1 = 35.6354
n2 = 82.8192
n3 = 3
print("n1 = {0:<10.4f}".format(n1), end="")         # n1 출력 4칸 공백
print("n2 = {0:<10.4f}".format(n2), end="")         # n2 출력 4칸 공백
print("n3 = {}\n".format(n3))                       # n3 출력 및 한줄 띄우기

print("n1+n2 = {0:>10.2f}".format(n1+n2), end="")   # 총 10칸 & 소수점 둘째자리까지 n1+n2 출력
print("{0:<4}".format(" "), end="")                 # 4칸 공백
print("n1-n2 ={0:>10.2f}".format(n1-n2))            # 총 10칸 & 소수점 둘째자리까지 n1-n2 출력


print("n1*n2 = {0:>10.2f}".format(n1*n2), end="")   # 총 10칸 & 소수점 둘째자리까지 n1*n2 출력
print("{0:<4}".format(" "), end="")                 # 4칸 공백
print("n1/n2 ={0:>10.2f}".format(n1/n2))            # 총 10칸 & 소수점 둘째자리까지 n1/n2 출력

print("n1**n3 ={0:>10.2f}".format(n1**n3), end="") # 총 10칸 & 소수점 둘째자리까지 n1**n3 출력
print("{0:<4}".format(" "), end="")                 # 4칸 공백
print("n1%n2 ={0:>10.2f}".format(n1%n2))            # 총 10칸 & 소수점 둘째자리까지 n1%n2 출력

print("n2//n1 ={0:>10.2f}".format(n2//n1), end="") # 총 10칸 & 소수점 둘째자리까지 n2//n1 출력
```

```python
# 4번 문제
import datetime
now = datetime.datetime.now() # 현재시간을 받아 now 에 저장
print("현재 시각은 :",now)
month = now.month             # month 라는 변수에 now.month(현재의 달) 저장
print("현재는 {}월 입니다.".format(month))
if month >= 3 and month <= 5:       # 3월~5월 사이일 경우
    print("현재는 봄입니다.")
elif month >= 6 and month <= 8:    # 6월~8월 사이일 경우
    print("현재는 여름입니다.")
elif month >= 9 and month <= 11:   # 9월~11월 사이일 경우
    print("현재는 가을입니다.")
else:                               # 그 외의 나머지 - (12월~2월)
    print("현재는 겨울입니다.")    
```

```python
# 5번 문제
money = 10000   # 적립된 금액 10000으로 초기화
print("10000원의 금액이 적립되어있습니다.")
prompt = """
다음 메뉴에 해당하는 번호를 입력하세요.
1. 아메리카노 1500
2. 카페라떼 2000
3. 에스프레소 1800
4. 주문을 끝냅니다.
"""

while money:   # money 는 항상 True 이므로 무한루프 
    order = int(input(prompt)) # order에 사용자가 입력한 값을 저장
    
    if (order == 1):    # 사용자가 1을 입력할 경우
        money = money - 1500                             
        print("손님께서는 아메리카노를 주문하셨습니다.")
        print("잔돈은 {}".format(money))
            
    elif (order == 2): # 사용자가 2를 입력할 경우
        money = money - 2000
        print("손님께서는 카페라떼를 주문하셨습니다.")
        print("잔돈은 {} 입니다.".format(money))
            
    elif (order == 3): # 사용자가 3을 입력할 경우
        money = money - 1800
        print("손님께서는 에스프레소를 주문하셨습니다.")
        print("잔돈은 {} 입니다.".format(money))
 
    elif (order == 4): # 사용자가 4를 입력할 경우
        print("주문을 끝냅니다.")
        break           
        
    else:                # 그 외의 숫자를 입력할 경우
        print("숫자를 잘못 입력하였습니다.")
        
    if money < 1500:
        print("잔돈은 {} 입니다.".format(money))
        break
        
```

```python
money = 10000

print(money, "의 금액이 적립되어 있습니다.")

prompt = """
 다음 메뉴에 해당하는 번호를 입력하세요.
 1. 딸기쥬스         2000
 2. 바나나쥬스       3000
 3. 주문을 끝냅니다.   
 4. 적립금 추가 
 """

while money:
    order = int(input(prompt))
    
    if(order == 1):
        money = money - 2000
        print("손님께서는 딸기쥬스를 주문하셨습니다.")
        print("잔돈은 {} 입니다.".format(money))
        
    if(order == 2):
        money = money - 3000
        print("손님께서는 바나나 쥬스를 주문하셨습니다.")
        print("잔돈은 {} 입니다.".format(money))
        
    if(order == 3):
        print("주문을 끝냅니다.")
        break
    
    if(order ==4):
        if(money < 2000):
            print("현재 적립금이 {} 입니다.".format(money))
            want = input("추가 적립을 원하십니까?")
            if(want == 'Y'):
                money = money + 10000
                print("적립금이 10000원 추가되어 현재 적립금은 {} 입니다.".format(money))
            if(want == 'N'):
                print("이용해주셔서 감사합니다.")
                break
```
