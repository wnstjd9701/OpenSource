지역별 전체 주문 횟수비교

```python
plt.figure(figsize=(10,5))
plt.bar(df['지역'].value_counts().reset_index()['index'],df['지역'].value_counts().reset_index()['지역'])
plt.xticks(rotation=30)
plt.title("지역별 주문횟수")
plt.show()
```

지역별 업종의 갯수 주문 횟수 확인

```python
df2 = df.groupby(['지역','업종']).count()['주문 건수'].reset_index()

for v in df2['지역'].unique():
    print("     ",v)
    df3 = df2.loc[df2['지역']==v].sort_values('주문 건수',ascending=False)
    plt.figure(figsize=(10,3))
    plt.bar(df3['업종'],df3['주문 건수'])
    plt.xticks(rotation=30)
    plt.show()
```

거의 대부분 지역구에서 카페/디저트를 가장 많이 주문했고, 세종특별자치시의 경우에는 카페/디저트를 제외한 다른 업종의 주문이 없다. 카페/디저트 주문 다음으로는 '치킨','한식'을 1순위로 시켰다. 해당 지역의 특성이라고 볼수 있다

서울 지역의 업종별 날짜에 따른 주문 횟수 변화확인

```python
df_se = df.loc[df['지역']=='서울특별시']
```

```python
df_count = df_se.groupby(['날짜','업종']).count()['주문 건수'].reset_index()
```

```python
df_count
```

```python
for value in df_count['업종'].unique():
    dfdf = df_count.loc[df_count['업종']==value]
    plt.figure(figsize=(10,3))
    plt.title(value)
    plt.plot(dfdf['날짜'],dfdf['주문 건수'])
    plt.show()
```

시간대별 업종별 주문 건수에 대해서 확인

```python
df_time = df.groupby(['시간대별 시간','업종']).sum().reset_index()
```

```python
df_time
```

```python
for value in df_time['시간대별 시간'].unique():
    df_t = df_time.loc[df_time['시간대별 시간']==value].sort_values('주문 건수')
    plt.figure(figsize=(10,3))
    plt.bar(df_t['업종'],df_t['주문 건수'])
    plt.title(str(value)+'시간의 업종별 주문 건수')
    plt.xticks(rotation=30)
    plt.show()
```
