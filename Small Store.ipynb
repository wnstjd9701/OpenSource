```python
import csv
from datetime import date

f = open('data.csv', 'r')
rdr = csv.reader(f)

day_total_by_cashier = {1: {}, 2:{}, 3: {}}
week_total_by_cashier = {1: {}, 2: {}, 3: {}}

def get_week_of_day(input_date):
    first_day_of_month = date(year=input_date.year, month=input_date.month, day=1)
    return (input_date.day + first_day_of_month.weekday() - 1) // 7 + 1
        
for line in rdr:
    order_date = date.fromisoformat(line[1])
    week = get_week_of_day(order_date)
    cachier, price, amount = int(line[2]), int(line[5]), int(line[6])
    total = week_total_by_cashier.get(cachier).get(week, 0) 
    
    if total == 0: 
        week_total_by_cashier.get(cachier)[week] = price * amount   
    else: 
        week_total_by_cashier.get(cachier)[week] += price * amount   
    
    
print(week_total_by_cashier)
f.close()
```
